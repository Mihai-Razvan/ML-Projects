```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings

warnings.filterwarnings('ignore')
```

```python
train_df = pd.read_csv('dataset/train.csv')
train_df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(train_df.drop(columns=['id', 'Heart Disease']), train_df['Heart Disease'])
```

```python
series = X_train['Age']

print('===Describe===')
print(series.describe())

print('\n===Unique Values===')
print(len(series.unique()))

print('\n===Percentages With Heart Disease By Value===')
percentages = []
for value in np.sort(series.unique()):
  mask = (series == value)
  percentage = (Y_train[mask] == 'Presence').sum() / Y_train[mask].count() * 100
  percentages.append(percentage)
  print(f'For {value}: {percentage:.2f}%')

plt.figure(figsize=(8, 8))
plt.title('Percentage With Heart Disease By Value')
plt.hist(percentages, bins=len(percentages))
plt.show()

plt.figure(figsize=(8, 8))
plt.title('Count Of Total Participants By Values')
plt.hist(series, bins=7)
plt.show()
```

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler

columns_to_remove = [

]
X_train = X_train.drop(columns=columns_to_remove)
X_test= X_test.drop(columns=columns_to_remove)

categorical_columns = [
]

label_encoders = {}
for col in categorical_columns:
  label_encoder = LabelEncoder()
  X_train[col] = label_encoder.fit_transform(X_train[col])
  X_test[col] = label_encoder.transform(X_test[col])
  label_encoders[col] = label_encoder

test_label_encoder = LabelEncoder()
Y_train = pd.Series(test_label_encoder.fit_transform(Y_train), index=Y_train.index, name='Heart Disease')
Y_test = pd.Series(test_label_encoder.transform(Y_test), index=Y_test.index, name='Heart Disease')

standard_scalers = {}
for col in X_train.columns:
  standard_scaler = StandardScaler()
  X_train[col] = standard_scaler.fit_transform(np.array(X_train[col]).reshape(-1, 1))
  X_test[col] = standard_scaler.transform(np.array(X_test[col]).reshape(-1, 1))
  standard_scalers[col] = standard_scaler

print(X_train.head())
print(Y_train.head())
```

```python
from sklearn.feature_selection import mutual_info_classif
import seaborn as sns

mi = mutual_info_classif(X_train, Y_train)
mi_scores = pd.Series(mi, index=X_train.columns).sort_values(ascending=False)
print(mi_scores)

plt.figure(figsize=(10, 6))
sns.barplot(x=mi_scores.values, y=mi_scores.index, palette="viridis")
plt.xlabel('Mutual Information')
plt.ylabel('Feature')
plt.title('Feature Importance by Mutual Information')
plt.tight_layout()
plt.show()
```
